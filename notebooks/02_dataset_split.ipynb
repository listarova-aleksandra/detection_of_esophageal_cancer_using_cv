```python
import pandas as pd
from sklearn.model_selection import train_test_split
from pathlib import Path
```

```python
df = pd.read_csv("C:/Users/user/Documents/esophageal_cancer_cv/data/processed/metadata.csv")
df
```

```python
df["label"].value_counts(normalize=True)
```

```python
train_df, temp_df = train_test_split(
    df,
    test_size=0.30,
    stratify=df["label"],
    random_state=42
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    stratify=temp_df["label"],
    random_state=42
)
```

```python
def check_split(name, split_df):
    print(f"{name}:")
    print("  size:", len(split_df))
    print("  label distribution:")
    print(split_df["label"].value_counts(normalize=True))
    print()

check_split("Train", train_df)
check_split("Val", val_df)
check_split("Test", test_df)
```

```python
Path("C:/Users/user/Documents/esophageal_cancer_cv/data/splits").mkdir(parents=True, exist_ok=True)
train_df.to_csv("C:/Users/user/Documents/esophageal_cancer_cv/data/splits/train.csv", index=False)
val_df.to_csv("C:/Users/user/Documents/esophageal_cancer_cv/data/splits/val.csv", index=False)
test_df.to_csv("C:/Users/user/Documents/esophageal_cancer_cv/data/splits/test.csv", index=False)
```
